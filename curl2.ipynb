```python
%matplotlib notebook
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

Plot the vector function:
\begin{equation}
\vec{F}(x,y) = y^2 \hat{x} + x^2\hat{y}
\end{equation}

```python
fig = plt.figure()
ax = fig.gca()

x, y = np.meshgrid(np.arange(-1, 1.2, 0.2),
                      np.arange(-1, 1.2, 0.2))

u = y**2
v = x**2 
w = 0
ax.quiver(x, y, u, v, scale=5)
plt.show()
ax.set_xlabel('X');
ax.set_ylabel('Y');
```

### As a demonstration use *sympy* to evaluate the curl (even though we can easily do it by hand)

```python
from sympy.physics.vector import ReferenceFrame
from sympy.physics.vector import curl
R = ReferenceFrame('R')

F = R[1]**2 * R.x + R[0]**2 * R.y
print('The vector field:')
print(F)
G = curl(F, R)  
print('\n')
print('The curl of the vector field')
print(G)
```
